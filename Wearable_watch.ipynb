<h3>2. Acquire training data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('survey 605.xlsx')
df.head(2)
```

Data is already clean, so just make shorten the columns name.

```python
df.isnull().sum()
```

```python
df1 =df.rename(columns={'Timestamp':'Timestamp', 'What is your age?':'Age', 'What is your gender?':"Gender",
       'What is your highest level of education?':'Education',
       'What is your current occupation?':'Ocupation',
       'How often do you exercise in a week?':'Excercise',
       'How long have you been using a fitness wearable?':'Period for usage',
       'How frequently do you use your fitness wearable?':'Frequency for usage',
       'How often do you track fitness data using wearable?':'Track data',
       'How has the fitness wearable impacted your fitness routine?':'Routine',
       'Has the fitness wearable helped you stay motivated to exercise?':'Motivated',
       'Do you think that the fitness wearable has made exercising more enjoyable?':'Enjoyable','How engaged do you feel with your fitness wearable?':'Engage',
       'Does using a fitness wearable make you feel more connected to the fitness community?':'Connected',
       'How has the fitness wearable helped you achieve your fitness goals?':'Goal',
       'How has the fitness wearable impacted your overall health?':'Impact',
       'Has the fitness wearable improved your sleep patterns?':'Sleep',
       'Do you feel that the fitness wearable has improved your overall well-being?':'Improved',
       'Has using a fitness wearable influenced your decision? [To exercise more?]':'To exercise more',
       'Has using a fitness wearable influenced your decision? [To purchase other fitness-related products?]':'purchase',
       'Has using a fitness wearable influenced your decision? [To join a gym or fitness class?]':'Gym',
       'Has using a fitness wearable influenced your decision? [To change your diet?]':'Diet'})
```

<h3>4. Analyze, identify patterns, and explore the data.

At first I want to check what kind of customers are purchasing this product.

```python
fig, g = plt.subplots()
g.pie(df1['Age'].value_counts(), labels=df1['Age'].value_counts().index, autopct='%1.1f%%')
g.set_title('Age Distribution')

plt.show()
```

```python
fig, g = plt.subplots()
g.pie(df1['Gender'].value_counts(), labels=df1['Gender'].value_counts().index, autopct='%1.1f%%')
g.set_title('Gender Distribution')

plt.show()
```

```python
fig, g = plt.subplots()
g.pie(df1['Education'].value_counts(), labels=df1['Education'].value_counts().index, autopct='%1.1f%%')
g.set_title('Education')

plt.show()
```

```python
fig, g = plt.subplots()
g.pie(df1['Ocupation'].value_counts(), labels=df1['Ocupation'].value_counts().index, autopct='%1.1f%%')
g.set_title('Ocupation')

plt.show()
```

Findings for customers:
1. Main customers were less than 34 Years(around 76%)
2. Main customers were more than collegue degree.(96.7%)
3. Main customers were employed(56.7%) and student(30%) 

Nextly I will check the corrlation between each questions.

```python
fig, g = plt.subplots()
g.pie(df1['Excercise'].value_counts(), labels=df1['Excercise'].value_counts().index, autopct='%1.1f%%')
g.set_title('Excercise')

plt.show()
```

```python
grid = sns.FacetGrid(df1, row="Frequency for usage", height=2.2, aspect=5)
grid.map_dataframe(sns.countplot, 'Excercise')
grid.fig.subplots_adjust(top=0.9)
grid.fig.suptitle('How many times doing excercise vs using wearables')
grid.set_axis_labels('How often Excersie', 'Times')
grid.add_legend()
```
